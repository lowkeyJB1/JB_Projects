```python
import random

## Generates a random word from our bank##
def word_generator():
    bank = ['frog', 'eagle', 'zebra', 'tiger', 'shark', 'dolphin', 'whale', 'spider',
             'lemur', 'ant', 'hippo', 'armadillo', 'axolotl ', 'babirusa', 'mongoose', 'cheetah', 'seadragon', 'blobfish'
            , 'octopus','snail', 'rhino', 'elephant', 'snake', 'stingray']
    word = random.choice(bank)
    return word

## updates the basic graphic for every right or wrong input ##
def graphic(basic_graphic, m_letter, c_letters, correct_animal):
    print(basic_graphic[len(m_letter)])
    print()
    print('Wrong Guesses:', end=' ')
    for l in m_letter:
        print(l, end=' ')
    print("\n")

    blanks = '_' * len(correct_animal)

    for i in range(len(correct_animal)):  # replaces the blanks with the right letters
        if correct_animal[i] in c_letters:
            blanks = blanks[:i] + correct_animal[i] + blanks[i+1:]

    for l in blanks:  # shows the correct animal 
        print(l, end=' ')
    print("\n")
    
## Creates the basic graphic for hangman###
basic_graphic = ["""
Guess the correct animal and he goes free!!!

   +---+
   |   |
       |
       |
       |
       |
       |
--------""", """
I'm Innocent!

  +---+
  |   |
  O   |
      |
      |
      |
      |
-------""", """
Help me!

  +---+
  |   |
  O   |
  |   |
      |
      |
      |
-------""", """
It's not that hard!

  +---+
  |   |
  O   |
 /|   |
      |
      |
      |
-------""", """
You're cutting in close

  +---+
  |   |
  O   |
 /|\  |
      |
      |
      |
-------""", """
Please!!!

  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
      |
-------""", """


  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
      |
-------"""]

## takes user input ###
def receive_guess(old_answer):
    while True:
        guess = input('Please guess a letter: ') ##user input ##
        guess = guess.lower()
        if len(guess) != 1:
            print('Please enter a letter.')
        elif guess in old_answer:
            print('You already guessed that letter. Please choose a different letter.')
        elif guess not in 'abcdefghijklmnopqrstuvwxyz':
            print('Please enter a letter in the aplhabet.')
        else:
            return guess
## ask user if they want to play again ###
def new_game():
    return input("\nDo you want to play another game of hangman? ").lower().startswith('y')


## actual hangman game##
m_letter = ''
c_letters = ''
correct_animal = word_generator()
finish = False
while True:
    graphic(basic_graphic, m_letter, c_letters, correct_animal)

    guess = receive_guess(m_letter + c_letters)

    if guess in correct_animal:
        c_letters = c_letters + guess

        foundAllLetters = True
        for i in range(len(correct_animal)):
            if correct_animal[i] not in c_letters:
                foundAllLetters = False
                break
        if foundAllLetters:
            print('\nYay!!! The The correct animal is "' +
                  correct_animal + '"! You saved him!')
            finish = True
    else:
        m_letter = m_letter + guess

        if len(m_letter) == len(basic_graphic) - 1:
            graphic(basic_graphic, m_letter,
                         c_letters, correct_animal)
            print('You have run out of guesses, he DIED!\nAfter ' + str(len(m_letter)) + ' wrong letters and ' +
                  str(len(c_letters)) + ' right letter, the animal was "' + correct_animal + '"')
            finish = True

    if finish:
        if new_game():
            m_letter = ''
            c_letters = ''
            finish = False
            correct_animal = word_generator()
        else:
            break
```

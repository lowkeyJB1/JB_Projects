```python
from random import randint

#create a list of hands
hand = ["Rock", "Paper", "Scissors"]

#gives the bot a random hand
bot = hand[randint(0,2)]
human_win = 0
bot_win = 0
ties = 0

## Actual Game"
human = False
while human == False:
    human = input("Enter Rock, Paper, Scissors, or STOP to exit them game! Please check spelling!: ")
    if human == bot:
        print("Tie!")
        ties +=1
    elif human == "Rock":
        if bot == "Paper":
            print('                                 ')
            print("Take your L!!!!", bot, "covers", human)
            print('                                 ')
            bot_win +=1
        else:
            print('                                 ')
            print("Victory!!!", human, "smashes", bot)
            print('                                 ')
            human_win +=1
    elif human == "Paper":
        if bot == "Scissors":
            print('                                 ')
            print("Here's an L!!!", bot, "cut", human)
            print('                                 ')
            bot_win +=1
        else:
            print('                                 ')
            print("Victory!!!", human, "covers", bot)
            print('                                 ')
            human_win +=1
    elif human == "Scissors":
        if bot == "Rock":
            print('                                 ')
            print("All aboad the L train!!!", bot, "smashes", human)
            print('                                 ')
            bot_win +=1
        else:
            print('                                 ')
            print("Victory!!!", human, "cuts", bot)
            print('                                 ')
            human_win +=1
    elif human == 'STOP':
        print('                                 ')
        print('You won', human_win, 'games, the bot won', bot_win, 'games, and there were', ties, 'tie or ties!!!!')
        print('                                 ')
        break
    else:
        print('                                 ')
        print("Please input Rock, Paper, or Scissors! Be wary of your spelling!: ")
        print('                                 ')
    human = False
    bot = hand[randint(0,2)]
```
